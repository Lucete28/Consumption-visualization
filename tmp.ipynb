```python
import os

def get_files_in_folder(folder_path):
    files = []
    for file_name in os.listdir(folder_path):
        full_path = os.path.join(folder_path, file_name)
        if os.path.isfile(full_path):
            files.append(full_path)
    return files

# 폴더 경로 설정
folder_path = './record/'

# 폴더 안에 있는 파일들 가져오기
files_in_folder = get_files_in_folder(folder_path)

# 결과 출력
print(files_in_folder)
```

```python
import pandas as pd
import win32com.client
import pyautogui
import time

# Excel 암호 입력 함수
def input_excel_password(password):
    pyautogui.typewrite(password)
    pyautogui.press('enter')

# Excel 앱 열기
excel_app = win32com.client.Dispatch("Excel.Application")

# 파일 경로 및 암호 설정
file_path = r"C:\PlayData\kakao_USed\record\카카오뱅크_거래내역_N6317902420_2023111015494237.xlsx"
file_password = "000303"

# Excel 파일 열기
workbook = excel_app.Workbooks.Open(file_path)

# Excel 암호 입력 (자동화)
input_excel_password(file_password)

# 시간을 주어 충분한 대기
time.sleep(2)

# 원하는 작업 수행
sheet = workbook.Sheets("카카오뱅크 거래내역")
df = pd.DataFrame(sheet.UsedRange())

# Excel 앱 종료
excel_app.Quit()
```
